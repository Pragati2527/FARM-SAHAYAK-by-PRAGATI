```python
##FARM SAHAYAK BY Pragati Chauhan

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Sample data: Yearly percentage of farmers using digital tools
data = {
    "Year": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    "Adoption_Rate": [5, 7, 10, 15, 22, 30, 40, 52, 65]  # in percentage
}

df = pd.DataFrame(data)
df.set_index("Year", inplace=True)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(df, marker='o')
plt.title("Adoption Rate of Digital Tools in Agriculture (2015-2023)")
plt.xlabel("Year")
plt.ylabel("Adoption Rate (%)")
plt.grid(True)
plt.show()

# Forecasting using Exponential Smoothing
model = ExponentialSmoothing(df, trend='add', seasonal=None, seasonal_periods=None)
fit = model.fit()
forecast = fit.forecast(5)  # Forecasting for the next 5 years

# Plotting the forecast
plt.figure(figsize=(10, 5))
plt.plot(df, marker='o', label='Historical')
plt.plot(forecast, marker='x', label='Forecast', color='red')
plt.title("Forecasted Adoption Rate of Digital Tools in Agriculture (2024-2028)")
plt.xlabel("Year")
plt.ylabel("Adoption Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Assuming initial revenue (A) based on 2023 adoption rate
initial_users = df["Adoption_Rate"].iloc[-1] / 100 * 10000000  # Assuming 10 million farmers
initial_revenue = initial_users * 500  # ₹500 per user per year

# Growth rate (B) based on historical data
growth_rate = 0.2  # Example growth rate

# Time (t) in years
years = np.arange(2024, 2029)

# Revenue projections
revenue_projections = initial_revenue * np.exp(growth_rate * (years - 2023))

# Assuming fixed costs (production, maintenance, marketing, etc.)
fixed_costs = 5000000  # Example fixed cost

# Profit projections
profit_projections = revenue_projections - fixed_costs

# Plotting the projections
plt.figure(figsize=(10, 5))
plt.plot(years, revenue_projections, marker='o', label='Revenue')
plt.plot(years, profit_projections, marker='x', label='Profit', color='green')
plt.title("Revenue and Profit Projections (2024-2028)")
plt.xlabel("Year")
plt.ylabel("Amount (₹)")
plt.legend()
plt.grid(True)
plt.show()

# Display the projections
projections = pd.DataFrame({
    "Year": years,
    "Revenue": revenue_projections,
    "Profit": profit_projections
})
print(projections)
```
