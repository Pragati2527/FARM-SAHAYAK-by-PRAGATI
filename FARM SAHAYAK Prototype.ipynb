```python
knowledge_base = {
    "how to grow rice": "To grow rice, you need a flooded field and rice seeds. Prepare the soil by plowing and level it before planting the seeds.",
    "best fertilizer for wheat": "The best fertilizer for wheat is a balanced NPK (Nitrogen, Phosphorus, and Potassium) fertilizer.",
    "pest control for tomatoes": "Use organic insecticides like neem oil or set up physical barriers like nets to protect tomatoes from pests.",
}
```

```python
#FARM SAHAYAK by Pragati Chauhan 

import speech_recognition as sr
from googletrans import Translator
import pyttsx3

# Initialize the translator
translator = Translator()

# Initialize the speech recognition
recognizer = sr.Recognizer()

# Initialize text-to-speech engine
engine = pyttsx3.init()

def speech_to_text():
    with sr.Microphone() as source:
        print("Please say something")
        audio = recognizer.listen(source)
        try:
            text = recognizer.recognize_google(audio)
            print(f"You said: {text}")
            return text
        except sr.UnknownValueError:
            print("Sorry, I did not understand that.")
            return None

def translate_text(text, src_lang='auto', dest_lang='en'):
    translated = translator.translate(text, src=src_lang, dest=dest_lang)
    return translated.text

def text_to_speech(text, lang='en'):
    engine.setProperty('voice', lang)
    engine.say(text)
    engine.runAndWait()

def get_response(query):
    for key in knowledge_base.keys():
        if key in query.lower():
            return knowledge_base[key]
    return "Sorry, I don't have information on that topic."

# Main flow
if __name__ == "__main__":
    # Step 1: Speech to Text
    query = speech_to_text()
    if query:
        # Step 2: Translate to English
        translated_query = translate_text(query, dest_lang='en')
        print(f"Translated Query: {translated_query}")
        
        # Step 3: Get response from knowledge base
        response = get_response(translated_query)
        print(f"Response: {response}")
        
        # Step 4: Translate response back to local language (if needed)
        translated_response = translate_text(response, src_lang='en', dest_lang='hi')  # Example: translating back to Hindi
        print(f"Translated Response: {translated_response}")
        
        # Step 5: Text to Speech
        text_to_speech(translated_response, lang='hi')  # Example: using Hindi voice
```
